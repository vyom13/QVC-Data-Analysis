```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df3 = pd.read_csv('C:/Users/shrey/Desktop/ABA Project/Group Project Option 2 QVC/social.csv')
df3.head(10)
```

```python
sentiment = df3['SENTIMENT']

df3['SENTIMENT'] = df3['SENTIMENT'].map({'Positive': 4, 'Neutral':3, 'Negative':2,'Strongly negative':1,'Strongly positive':5})
```

```python
df3.head(10)
```

```python
df3['INTERACTION_DATE'] = pd.to_datetime(df3['INTERACTION_DATE'])
df3['INTERACTION_DATE'] = df3['INTERACTION_DATE'].dt.date

df3.head(10)
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
df3 = df3['SENTIMENT'].groupby(df3['INTERACTION_DATE']).agg(['mean']).reset_index()
```

```python
df3
```

```python
df3 = df3.rename(columns={'INTERACTION_DATE': 'DATE'})
df3.head(10)
```

```python
df3 = df3.rename(columns={'mean': 'SCORE'})
df3.head(10)
```

```python
header = ["DATE", "SCORE"]
df3.to_csv('data_2.csv', columns = header)
```
